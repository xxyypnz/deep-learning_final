# PostgreSQL SQL 测试用例生成作业

请根据 `data/test_v3.json` 中的 PostgreSQL commit 信息，生成 `submission.json`。

最终提交文件应保存到：

```text
outputs/submission.json
```

提交 JSON 的每条记录需要包含 `id` 和 `generated_sql_tests`，其中 SQL 放在 `<sql>...</sql>` 标签内。

````python
from pathlib import Path
import json

DATA_DIR = Path('data')
OUTPUT_DIR = Path('outputs')
OUTPUT_DIR.mkdir(exist_ok=True)

dataset_path = DATA_DIR / 'test_v3.json'
with dataset_path.open(encoding='utf-8') as f:
    test_data = json.load(f)

len(test_data), test_data[0].keys()
````

下面是一个最小格式示例。正式提交时，你需要为更多 commit 生成更有针对性的 SQL 测试用例。

````python
submission = []

for item in test_data[:1]:
    submission.append({
        'id': item['id'],
        'generated_sql_tests': '''<test_cases>
  <test_case id="1">
    <description>最小示例 SQL，请替换为针对 commit 的测试。</description>
    <sql>
SELECT version();
    </sql>
  </test_case>
</test_cases>'''
    })

output_path = OUTPUT_DIR / 'submission.json'
with output_path.open('w', encoding='utf-8') as f:
    json.dump(submission, f, ensure_ascii=False, indent=2)

output_path
````
